```python
def convert(text):
     l=text.split(',')
     l.pop()
     return l
```

```python
movies['Genere']=movies['Genere'].apply(convert)
```

```python
movies.head(2)
```

```python
def convert_description(text):
       l=text.split()
       return l
```

```python
movies['Description']=movies['Description'].apply(convert_description)
```

```python
movies.head(2)
```

```python
movies['tags']=movies['Description']+movies['Genere']
```

```python
new_df=movies[['Rank','Movie Names','tags']]
```

```python
new_df.head(2)
```

```python
def l2s(list):
     s=' '.join(list)
     return s
```

```python
new_df['tags']=new_df['tags'].apply(l2s)
```

```python
new_df.head(2)
```

```python
new_df.iloc[0]['tags']
```

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
ps=PorterStemmer()
```

```python
def stems(text):
   l=[]
   for i in text.split():
       l.append(ps.stem(i))
   return ' '.join(l)
```

```python
new_df['tags']=new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=300,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['Movie Names']=='Dangal'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['Movie Names']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x: x[1])
    for i in distances[1:11]:
           print(new_df.iloc[i[0]]['Movie Names'])
```

```python
recommend('Major')
```
